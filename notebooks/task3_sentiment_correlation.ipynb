```python
# 1. Load data
import pandas as pd
from src import sentiment, correlation

df = pd.read_csv("data/processed/news_stock_merged.csv", parse_dates=['date_utc_date'])

# 2. Compute sentiment
df = sentiment.compute_vader_sentiment(df)
df = sentiment.compute_transformer_sentiment(df)

# 3. Aggregate daily sentiment
daily_sent = sentiment.aggregate_daily_sentiment(df)

# 4. Merge with stock returns
# Assuming stock_df has columns: date_utc_date, daily_return
stock_df = pd.read_csv("data/processed/stock_returns.csv", parse_dates=['date_utc_date'])
merged_df = correlation.merge_sentiment_returns(daily_sent, stock_df)

# 5. Compute correlations
sentiment_cols = [col for col in merged_df.columns if 'sentiment' in col]
corr_results = correlation.compute_correlations(merged_df, sentiment_cols)
print(corr_results)

# 6. Optional: scatter plot for visual check
correlation.plot_correlation(merged_df, sentiment_cols[0], 'daily_return')
```
